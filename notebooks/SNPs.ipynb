Overlap with 3 methods
--------------------------

For those strains for which we have the three methods available, how much do they overlap?

```python
d = {}
```

```python
%%cache SNPs_1.pkl d
for f in mfiles.intersection(pfiles).intersection(kfiles):
    d[f] = (set(), set(), set())
    for r in vcf.Reader(open('../mout/%s.vcf'%f)):
        if not r.is_indel:
            d[f][0].add( (r.POS, str(r.ALT[0])) )
        else:
            if len(str(r.REF)) == len(str(r.ALT[0])):
                # Multiple SNP!
                for ref, alt, pos in zip(str(r.REF), str(r.ALT[0]), range(r.POS, r.POS+len(str(r.REF)))):
                    if ref != alt:
                        d[f][0].add( (pos, alt) )
    for r in vcf.Reader(open('../pout/%s.vcf'%f)):
        if not r.is_indel == 'SNV':
            d[f][1].add( (r.POS, str(r.ALT[0])) )
    for r in vcf.Reader(open('../kout/%s.vcf'%f)):
        if not r.is_indel == 'SNV':
            d[f][2].add( (r.POS, str(r.ALT[0])) )
```

```python
subsets = ['intersection',
           'map/parsnp',
           'map/ksnp',
           'parsnp/ksnp',
           'map',
           'parsnp',
           'ksnp']
```

```python
res = []
for sample, v in d.items():
    res.append( {'#': (len(v[0].intersection(v[1]).intersection(v[2])),
                        len(v[0].intersection(v[1]) - v[2]),
                        len(v[0].intersection(v[2]) - v[1]),
                        len(v[1].intersection(v[2]) - v[0]),
                        len(v[0] - v[1].union(v[2])),
                        len(v[1] - v[0].union(v[2])),
                        len(v[2] - v[1].union(v[0]))),
                 'sample': [sample for x in range(7)],
                 'subset': subsets} )
```

```python
snp3 = pd.concat([pd.DataFrame(x) for x in res])
```

```python
plt.figure(figsize=(12, 7))

sns.boxplot([snp3[snp3['subset'] == x]['#'] for x in subsets],
           names=subsets,
           color=(sns.xkcd_rgb["dark grey"],
                   sns.xkcd_rgb["grey"],
                   sns.xkcd_rgb["grey"],
                   sns.xkcd_rgb["grey"],
                   sns.xkcd_rgb["light grey"],
                   sns.xkcd_rgb["light grey"],
                   sns.xkcd_rgb["light grey"]))

plt.xlabel('Subset')
plt.ylabel('Number of SNPs')
plt.title('Samples with three methods overlapping: %d'%len(
        mfiles.intersection(pfiles).intersection(kfiles)))

sns.despine()
```

Coding/non-coding ratio
---------------------------

```python
coding = [(int(f.location.start), int(f.location.end))
          for f in s.features
          if f.type == 'CDS' and
          'pseudo' not in f.qualifiers]
```

```python
%%cache SNPs_6.pkl cod non_cod dcoding
# What is the proportion of coding/non-coding?
cod = 0 
non_cod = 0

dcoding = {}

for x in range(1, len(s)+1):
    b = False
    for s, e in coding:
        if s <= x <= e:
            b = True
            break
    if b:
        cod += 1
        dcoding[x] = True
    else:
        non_cod += 1
        dcoding[x] = False
```

```python
plt.figure(figsize=(7, 7))

s = SeqIO.read('../genome.gbk', 'genbank')
total = len(s)

def my_autopct(pct):
    val=int(pct*total/100.0)
    return '{v:d}'.format(v=val)

a=plt.pie([cod, non_cod], labels=['coding', 'non-coding'],
      explode=[0.1, 0], radius=0.9,
      colors=[(1, 0, 0, float(x)/total) for x in (cod, non_cod)],
      autopct=my_autopct)
```

```python
%%cache SNPs_7.pkl mcoding pcoding kcoding
mcoding= {}
for f in mfiles:
    mcoding[f] = [0, 0]
    for var in d[f][0]:
        if dcoding[var[0]]:
            mcoding[f][0] += 1
        else:
            mcoding[f][1] += 1
pcoding= {}
for f in pfiles:
    pcoding[f] = [0, 0]
    for var in d[f][1]:
        if dcoding[var[0]]:
            pcoding[f][0] += 1
        else:
            pcoding[f][1] += 1
kcoding= {}
for f in kfiles:
    kcoding[f] = [0, 0]
    for var in d[f][2]:
        if dcoding[var[0]]:
            kcoding[f][0] += 1
        else:
            kcoding[f][1] += 1
```

```python
plt.figure(figsize=(7, 7))

sns.violinplot([[mcoding[x][1]/float(mcoding[x][0]) for x in mcoding],
             [pcoding[x][1]/float(pcoding[x][0]) for x in pcoding],
             [kcoding[x][1]/float(kcoding[x][0]) for x in kcoding]],
           names=['map', 'parsnp', 'ksnp'],
           color=(sns.xkcd_rgb["dark grey"],
                   sns.xkcd_rgb["grey"],
                   sns.xkcd_rgb["light grey"]))

plt.hlines(float(non_cod)/cod, 0, 3.5, linestyles='dashed', colors=['r'])

plt.ylabel('non-coding/coding ratio')
plt.xlabel('method')
plt.title('Genome non-coding/coding ratio: %.2f'%(float(non_cod)/cod))

sns.despine()
```

Conserved/variable genes
----------------------------

```python
ls_bsr = pd.read_table('../lreference/bsr_matrix_values.txt')
```

```python
ls_bsr.rename(columns={ls_bsr.columns[0]:'Gene'}, inplace=True)
ls_bsr.set_index('Gene', inplace=True)
```

```python
conservation = ls_bsr.sum(axis=1)
```

```python
%%cache SNPs_8.pkl mgenes pgenes kgenes
mgenes = {}
for f in [x for x in os.listdir('../mout/') if x.endswith('.nonsyn.vcf')]:
    mgenes[f.replace('.nonsyn.vcf', '')] = {}
    for r in vcf.Reader(open('../mout/%s'%f)):
        g = r.INFO['CDS']
        mgenes[f.replace('.nonsyn.vcf', '')][g] = mgenes[f.replace('.nonsyn.vcf', '')].get(g, 0)
        mgenes[f.replace('.nonsyn.vcf', '')][g] += 1
pgenes = {}
for f in [x for x in os.listdir('../pout/') if x.endswith('.nonsyn.vcf')]:
    pgenes[f.replace('.nonsyn.vcf', '')] = {}
    for r in vcf.Reader(open('../pout/%s'%f)):
        g = r.INFO['CDS']
        pgenes[f.replace('.nonsyn.vcf', '')][g] = pgenes[f.replace('.nonsyn.vcf', '')].get(g, 0)
        pgenes[f.replace('.nonsyn.vcf', '')][g] += 1
kgenes = {}
for f in [x for x in os.listdir('../kout/') if x.endswith('.nonsyn.vcf')]:
    kgenes[f.replace('.nonsyn.vcf', '')] = {}
    for r in vcf.Reader(open('../kout/%s'%f)):
        g = r.INFO['CDS']
        kgenes[f.replace('.nonsyn.vcf', '')][g] = kgenes[f.replace('.nonsyn.vcf', '')].get(g, 0)
        kgenes[f.replace('.nonsyn.vcf', '')][g] += 1
```

```python
mallgenes = set()
for f, v in mgenes.items():
    for g in v:
        mallgenes.add(g)
pallgenes = set()
for f, v in pgenes.items():
    for g in v:
        pallgenes.add(g)
kallgenes = set()
for f, v in kgenes.items():
    for g in v:
        kallgenes.add(g)
```

```python
mvalues = []
for g in mallgenes:
    varg = []
    for f, v in mgenes.items():
        if g in v:
            varg.append(v[g])
    mvalues.append( (max(varg), conservation.loc[g]) )
pvalues = []
for g in pallgenes:
    varg = []
    for f, v in pgenes.items():
        if g in v:
            varg.append(v[g])
    pvalues.append( (max(varg), conservation.loc[g]) )
kvalues = []
for g in kallgenes:
    varg = []
    for f, v in kgenes.items():
        if g in v:
            varg.append(v[g])
    kvalues.append( (max(varg), conservation.loc[g]) )
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot([x[1] for x in mvalues],
         [x[0] for x in mvalues],
         'r.',
         alpha=0.3)

plt.ylabel('Max. number of SNPs per gene')
plt.xlabel('Gene conservation (sum of LS-BSR score)')
plt.title('map')

plt.subplot(1, 3, 2)
plt.plot([x[1] for x in pvalues],
         [x[0] for x in pvalues],
         'b.',
         alpha=0.33)
plt.xlabel('Gene conservation (sum of LS-BSR score)')
plt.title('parsnp')

plt.subplot(1, 3, 3)
plt.plot([x[1] for x in kvalues],
         [x[0] for x in kvalues],
        'g.',
        alpha=0.33)
plt.xlabel('Gene conservation (sum of LS-BSR score)')
plt.title('ksnp')

sns.despine()
```
